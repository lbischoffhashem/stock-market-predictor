```python
import yfinance as yf
sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=500, min_samples_split=100, random_state=1, n_jobs=-1)

# Percentage change from open to close (daily return)
sp500["OC_Pct_Change"] = (sp500["Close"] - sp500["Open"]) / sp500["Open"]

# Volatility (Range as a percentage of the open)
sp500["Daily_Volatility"] = (sp500["High"] - sp500["Low"]) / sp500["Open"]

# Where did the price close relative to its daily range?
sp500["Close_Relative_To_Range"] = (sp500["Close"] - sp500["Low"]) / (sp500["High"] - sp500["Low"])

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

#intial model ignoring volume for now.
predictors = ["OC_Pct_Change","Daily_Volatility","Close_Relative_To_Range"]

model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score
import pandas as pd


preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds],axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    probs = model.predict_proba(test[predictors])[:, 1]
    probs = pd.Series(probs, index=test.index, name="Predicted_Prob")
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds, probs],axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
from sklearn.metrics import f1_score
import numpy as np

def optimize_f1_score():
    best_f1 = 0
    best_thresh = 0
    
    for thresh in np.arange(0.25, 1.01, 0.01):
        preds = (predictions["Predicted_Prob"] >= thresh).astype(int)
        score = f1_score(predictions["Target"], preds, zero_division=0)
        if score > best_f1:
            best_f1 = score
            best_thresh = thresh
    return best_thresh, best_f1
```

```python
import numpy as np
from sklearn.metrics import fbeta_score

def optimize_fbeta_score():
    best_fbeta = 0
    best_beta = 0.15
    best_thresh = 0

    for thresh in np.arange(0.51, 1.01, 0.01):
        y_pred = (predictions["Predicted_Prob"] >= thresh).astype(int)
        score = fbeta_score(predictions["Target"], y_pred, beta=best_beta, zero_division=0)
        
        if score > best_fbeta:
            best_fbeta = score
            best_thresh = thresh
    return best_thresh, best_fbeta
```
